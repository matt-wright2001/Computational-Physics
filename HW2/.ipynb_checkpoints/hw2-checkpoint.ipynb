```python
import matplotlib.pyplot as plt
import numpy as np
from trapInt import funct, trap, trapIntMain
```

### Simpson's Rule

```python
def simp(n, min, max):
    if n % 2 == 0:
        print("Error: n must be odd")
        return None
        
    dx = (max - min) / (n-1)
    sum = 0.0

    for i in range(1, n-1):
        x = min + i*dx
        if i % 2 == 0:
            sum += 2*funct(x)
        else:
            sum += 4*funct(x)
    
    # Add the endpoints
    sum += funct(min) + funct(max)
    sum *= dx/3.0

    return sum
```

### Testing Simpson's Rule function

```python
a = 1.0
b = 2.0
exact = (2/3) * (b ** (3/2) - a ** (3/2))

for i in range(3, 10000, 2):
    approx = simp(i, a, b)
    error = abs(exact - approx)
    print('{:6d}{:.12e}'.format(i, error))
for i in range(1001, 3000000, 10000):
    approx = simp(i, a, b)
    error = abs(exact - approx)
    print('{:6d}{:.12e}'.format(i, error))
    
```

## Error plots

```python
trapezoid_errors = []
simpson_errors = []
ns = list(range(3, 10000, 2)) + list(range(10001, 3000000, 10000))

for i in ns:
    trap_error = abs(exact - trap(i, a, b))
    simp_error = abs(exact - simp(i, a, b))
    trapezoid_errors.append(trap_error)
    simpson_errors.append(simp_error)

# Plotting the log-log plot
plt.figure()
plt.loglog(ns, trapezoid_errors, label='Trapezoid Rule Error')
plt.loglog(ns, simpson_errors, label='Simpsons Rule Error')
plt.xlabel('N (Number of Points)')
plt.ylabel('Error')
plt.title('Error vs N')
plt.legend()
plt.show()

# Finding regression values
# Ignoring small values of N for regression by selecting a suitable start index
start_index = 50
x_log = np.log(ns[start_index:])
trap_log_error = np.log(trapezoid_errors[start_index:])
simp_log_error = np.log(simpson_errors[start_index:])

# Finding the regression parameters
trap_coeffs = np.polyfit(x_log, trap_log_error, 1)
simp_coeffs = np.polyfit(x_log, simp_log_error, 1)

print("Trapezoid Rule: Approximation error power law dependence: ", trap_coeffs[0])
print("Simpson's Rule: Approximation error power law dependence: ", simp_coeffs[0])
```
